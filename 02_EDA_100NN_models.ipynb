# Load plugs data

```python
url = 'https://raw.githubusercontent.com/sergioGarcia91/ML_and_EDA/main/Datasets/PlugsMesaLosSantos_2022.csv'
# CSV lugs
df_Plugs = pd.read_csv(url, sep=';', decimal='.')

df_Plugs.info()
```

# Models list

```python
models_list = os.listdir(path_save_models)
models_list = [model for model in models_list if model.endswith('.pkl')]
models_list
```

## Thesis model

```python
!wget https://github.com/sergioGarcia91/AplicativoMesaLosSantos/raw/refs/heads/main/modelosNN/porosidadNNrelu_hl25_8_5Sc32.pkl
```

```python
model_thesis = 'porosidadNNrelu_hl25_8_5Sc32.pkl'
model_thesis = joblib.load(model_thesis)

model_thesis
```

# Get data of interest

```python
df_Plugs.columns
```

```python
df = df_Plugs[['X', 'Y', 'Z',
               'Unidad', 'Distancia Falla [m]', 'Porosidad [%]', 'Litologia']].copy()

df.head()
```

# Predict with the thesis model

```python
df['Unidad'].unique()
```

```python
dict_geo = {'K1t':8,
            'K1p':7,
            'K1r':6,
            'K1ls_ms':5,
            'K1ls_mm':4,
            'K1ls_mi':3,
            'J1-2j':2,
            'J1gp':1,
            'Oss':0}

df['GeoNN'] = df['Unidad'].replace(dict_geo, inplace=False)

df.head()
```

## Data normalization thesis

```python
X_tesis = df[['X', 'Y', 'Z', 'Distancia Falla [m]', 'GeoNN']].copy()
X_tesis['Xn'] = (X_tesis['X'] - 1096700) / (1120100 - 1096700)
X_tesis['Yn'] = (X_tesis['Y'] - 1235500) / (1260100 - 1235500)
X_tesis['Zn'] = (X_tesis['Z'] - (-150)) / (1830 - (-150))
X_tesis['DFn'] = (X_tesis['Distancia Falla [m]'] - (000)) / (3000 - (000))

X_tesis = X_tesis[['Xn', 'Yn', 'Zn', 'DFn', 'GeoNN']].copy()

X_tesis.head()
```

```python
X_tesis = X_tesis.to_numpy()
```

```python
df['Porosity NN Thesis'] = np.round(model_thesis.predict(X_tesis), 2)

df.head()
```

# Predict with 100 models

```python
X_100models = df_Plugs[['X', 'Y', 'Z', 'Litologia', 'Rho Wet [Ohm*m]',]]

X_100models
```

## Data normalization 100 models



```python
X_100models['Rho Wet [Ohm*m]'] = np.log10(X_100models['Rho Wet [Ohm*m]'])

X_100models['X'] = (X_100models['X'] - 1096700) / (1120100 - 1096700)
X_100models['Y'] = (X_100models['Y'] - 1235500) / (1260100 - 1235500)
X_100models['Z'] = (X_100models['Z'] - (500)) / (1830 - (500))
X_100models['Rho Wet [Ohm*m]'] = (X_100models['Rho Wet [Ohm*m]'] - (1)) / (5 - (1))

X_100models.describe().round(2)
```

```python
X_100models  = pd.get_dummies(X_100models, columns=['Litologia'],
                             prefix='',
                             dtype=int)

X_100models.head()
```

```python
X_100models = X_100models.to_numpy()
```

```python
df_100models = pd.DataFrame()

for model in models_list:
  nn_model = model[6:12]
  model = joblib.load(path_save_models + model)
  df_100models[nn_model] = np.round(model.predict(X_100models), 2)

df_100models.head()
```

## Load DF models

```python
df_models = pd.read_csv(path_save_models + 'df_100models.csv', sep=';', decimal='.')
df_models = df_models.sort_values(by='Test score', ascending=False)

df_models.head() # Model 088 has the best test score
```

## Mean and Standard deviation

```python
df['Mean 100models'] = df_100models.mean(axis=1)
df['Std 100models'] = df_100models.std(axis=1)

df['Model 88'] = df_100models['NN_088']

df.head()
```

# Boxplot

```python
df['Litologia'].unique()
```

```python
replace_dict = {
    'arenisca': 'sandstone',
    'arenisca calcarea': 'calcareous sandstone',
    'biomicrita': 'biomicrite',
    'caliza': 'limestone',
    'conglomerado': 'conglomerate',
    'esquisto': 'schist',
    'granito': 'granite',
    'lodo calcareo': 'calcareous mudstone',
    'lodolita': 'lodolite',
    'metaarenisca': 'meta-sandstone'
}
```

```python
df['Lithology'] = df['Litologia'].replace(replace_dict, inplace=False)

df.head()
```

```python
Lithology_order = ['sandstone',
                   'calcareous sandstone',
                   'biomicrite',
                   'limestone',
                   'conglomerate',
                   'schist',
                   'granite',
                   'calcareous mudstone',
                   'lodolite',
                   'meta-sandstone'
                   ]

colors_lito = {'sandstone': 'yellow',
               'lodolite': 'lime',
               'conglomerate': 'tab:orange',
               'calcareous mudstone': 'tab:olive',
               'limestone': 'teal',
               'biomicrite': 'tab:cyan',
               'calcareous sandstone': 'tab:brown',
               'granite': 'tab:pink',
               'meta-sandstone': 'tab:green',
               'schist': 'tab:purple'}
```

```python
save_plots = True

for col in ['Porosity NN Thesis', 'Mean 100models', 'Model 88']:
  plt.figure(figsize=(14, 5))

  sns.boxplot(x='Lithology',
              y='Porosidad [%]',
              data=df,
              palette = colors_lito,
              order=Lithology_order)

  sns.swarmplot(x='Lithology',
                y=col,
                data=df,
                color='r',
                size=4.0,
                order=Lithology_order)

  plt.xticks(rotation=45)
  plt.grid(ls='--', color='grey', alpha=0.8)
  plt.ylabel('Porosity [%]')
  plt.xlabel('Lithology')

  if col == 'Porosity NN Thesis':
    plt.title(f'Porosity NN García-Arias  et al. (2024)')
  else:
    plt.title(col)

  if save_plots:
    col = col.replace(' ', '_')

    plt.savefig(path_save_plots + f'Boxplot_{col}.png',
                  dpi=500,
                  bbox_inches = 'tight',
                  pad_inches=0.25)

  plt.show()
  print('\n')
```

## Std

```python
plt.figure(figsize=(14, 5))

sns.boxplot(x='Lithology',
            y='Std 100models',
            data=df,
            palette = colors_lito,
            order=Lithology_order)


plt.xticks(rotation=45)
plt.grid(ls='--', color='grey', alpha=0.8)
plt.ylabel('Porosity [%] - Std')
plt.xlabel('Lithology')
plt.title('Porosity Standard Deviation - 100 Models')

if save_plots:
  plt.savefig(path_save_plots + f'Boxplot_Porosity_std.png',
                dpi=500,
                bbox_inches = 'tight',
                pad_inches=0.25)

plt.show()
```

```python
bins_poro = np.arange(0, 31, 1)

for lito in df['Lithology'].unique():
  mask_lito = df['Lithology'] == lito
  data_df = df[mask_lito]
  data_100models = df_100models[mask_lito]
  data_100models = pd.DataFrame(data_100models.to_numpy().reshape((-1,1)), columns=['Porosity'])

  fig, axs = plt.subplots(2, 2, figsize=(12, 7))
  sns.histplot(data=data_df,
               x='Porosidad [%]',
               bins=bins_poro,
               ax=axs[0,0],
               color='tab:blue')

  axs[0,0].grid(ls='--', color='grey', alpha=0.8)
  axs[0,0].set_title(f'Plugs')
  axs[0,0].set_xlabel('Porosity [%]')

  sns.histplot(data=data_100models,
               x='Porosity',
               bins=bins_poro,
               ax=axs[0,1],
               color='tab:orange')

  axs[0,1].grid(ls='--', color='grey', alpha=0.8)
  axs[0,1].set_title(f'100 models')
  axs[0,1].set_xlabel('Porosity [%]')

  sns.histplot(data=data_df,
               x='Porosity NN Thesis',
               bins=bins_poro,
               ax=axs[1,0],
               color='tab:green')

  axs[1,0].grid(ls='--', color='grey', alpha=0.8)
  axs[1,0].set_title(f'García-Arias  et al. (2024)')
  axs[1,0].set_xlabel('Porosity [%]')

  sns.histplot(data=data_df,
               x='Model 88',
               bins=bins_poro,
               ax=axs[1,1],
               color='tab:purple')

  axs[1,1].grid(ls='--', color='grey', alpha=0.8)
  axs[1,1].set_title(f'Model 88')
  axs[1,1].set_xlabel('Porosity [%]')

  plt.suptitle(f'Porosity distribution - {lito}')
  plt.tight_layout()

  plt.savefig(path_save_plots + f'Histograms_{lito}.png',
                dpi=500,
                bbox_inches = 'tight',
                pad_inches=0.25)

  plt.show()

  print('\n')
```
